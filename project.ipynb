```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
df.head(7)
```

```python
df.isna().sum()
```

```python
df = df.dropna(axis = 1)
```

```python
df.isna().sum()
```

```python
sns.countplot(df['diagnosis'], label='count')
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_Y = LabelEncoder()
df.iloc[:,1] = labelencoder_Y.fit_transform(df.iloc[:,1].values)
df.iloc[:,1]
```

```python
sns.pairplot(df.iloc[:,1:5],hue='diagnosis')
```

```python
df.iloc[:,1:12].corr()
```

```python
plt.figure(figsize = (10, 10))
sns.heatmap(df.iloc[:,1:12].corr(), annot = True, fmt = '.0%')
```

```python
X = df.iloc[:,2:31].values
Y = df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25 , random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
def models(x_train,y_train):
     from sklearn.linear_model import LogisticRegression
     log=LogisticRegression(random_state=0)
     log.fit(x_train,y_train)
     print("Logistic Regression Training Accuracy",log.score(x_train,y_train))
     return log
```

```python
model=models(X_train,Y_train)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,model.predict(X_test))
TP=cm[1][1]
TN=cm[0][0]
FN=cm[0][1]
FP=cm[1][0]
print(cm)
print("Testing Accuracy is ",(TP+TN)/(TP+TN+FP+FN))
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
print(classification_report(Y_test,model.predict(X_test)))
print(accuracy_score(Y_test,model.predict(X_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
forest.fit(X_train,Y_train)
```

```python
print("Random Forest Training Accuracy",forest.score(X_train,Y_train))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,forest.predict(X_test))
TP=cm[1][1]
TN=cm[0][0]
FN=cm[0][1]
FP=cm[1][0]
print(cm)
print("Testing Accuracy for random forest  is ",(TP+TN)/(TP+TN+FP+FN))
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
print(classification_report(Y_test,forest.predict(X_test)))
print(accuracy_score(Y_test,forest.predict(X_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(criterion='entropy',random_state=0)
tree.fit(X_train,Y_train)
```

```python
print("Decision Training Accuracy",tree.score(X_train,Y_train))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,tree.predict(X_test))
TP=cm[1][1]
TN=cm[0][0]
FN=cm[0][1]
FP=cm[1][0]
print(cm)
print("Testing Accuracy for decision Tree  is ",(TP+TN)/(TP+TN+FP+FN))
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
print(classification_report(Y_test,tree.predict(X_test)))
print(accuracy_score(Y_test,tree.predict(X_test)))
```
